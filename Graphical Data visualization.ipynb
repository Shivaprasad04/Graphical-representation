```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ['c','c+','java','python']
ex = [0.2,0.0,0.0,0.0]
c = ['r','b','y','g']

plt.pie(x,labels = y,explode = ex,colors = c,autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie([1],radius= 0.5)
plt.show()
```

```python
x = [10,20,30,40]
x1 = [20,30,40,50]
y = ['c','c+','java','python']


plt.pie(x,labels = y,radius = 1)
plt.pie([1],radius = 0.8,colors = 'w')
plt.show()
```

```python
x = [10,20,30,40]
x1 = [20,30,40,50]
y = ['c','c+','java','python']


plt.pie(x,labels = y,radius = 1)
cr = plt.Circle(xy = (0,0),radius = 0.6,facecolor = 'w')
plt.gca().add_artist(cr)

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
mtcars = pd.read_csv(r"C:\Users\shiva\Downloads\mtcars.csv")
mtcars.head()
```

```python
# bar plot
res = sns.barplot(x = 'cyl',y = 'mpg',data = mtcars,palette = 'Set1')
res
```

```python
# Count plot
sns.countplot(x = 'cyl' ,data = mtcars, palette = 'rocket')
```

```python
sns.countplot(y = 'carb',data = mtcars,palette = 'Set2')
```

```python
sns.countplot(x = 'gear',hue = 'cyl',data = mtcars)
```

```python
# Dist plot
sns.distplot(mtcars.mpg,bins = 10,color = 'b')
```

```python
# Heatmap
sns.heatmap(mtcars.corr(),cbar = True,linewidths = 0.2)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length',data = iris)
```

```python
sns.scatterplot(x = 'sepal_length', y = 'petal_length',data = iris,hue = 'species')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue = 'species',palette = 'Set1')
```

```python
#lm plot

sns.lmplot(x = 'petal_length',y = 'petal_width',data = iris)
```

```python
sns.lmplot(x = 'petal_length',y = 'petal_width',hue = 'species',data = iris,markers = ['*','+','o'])
```

```python
# Box plot

sns.boxplot(x = 'species',y = 'sepal_width',data = iris)
```
